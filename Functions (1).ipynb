```python
#. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum_even_numbers(my_list):
  sum = 0
  for number in my_list:
    if number % 2 == 0:
      sum += number
  return sum
sum_even_numbers([10,20,11,13,15,14,23,22])

```

```python
 #. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
 def square_of_numbers(my_list):
  square_of_the_list = []
  for number in my_list:
    number = int(number)
    square_of_the_list.append(number**2)
  return square_of_the_list
square_of_numbers([10,20,30,40,50])

```

```python
#. Write a Python function that checks if a given number is prime or not from 1 to 200.
def is_prime(number):
  if number > 1 and number <= 200:
    if number == 2 or number == 3 or number == 5 :
      print("it is prime")
    elif number % 2 == 0 or number % 3 == 0 or number % 5 == 0:
      print("it is not a prime")
    else :
      print("It is a prime number ")
  else:
    print("Please a Number from  1 to 200")
```

```python
#Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fibonacci_sequence(numbers):
  x , y = 0 ,1
  my_list = []
  for number in range(numbers):
    my_list.append(x)
    x , y = x + y , x
  return my_list
my_iterator = iter(fibonacci_sequence(10))
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
```

```python
#Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_2(numbers):
  my_list = []
  for number in range(numbers):
    result = number ** 2
    my_list.append(result)
  yield my_list
a = power_2(10)
for item in a:
  print(item,end = ' ')
```

```python
#Implement a generator function that reads a file line by line and yields each line as a string.
def gen_fun(my_function):
  length= len(my_function)
  i = 0
  while i <= length-1:
    yield str(my_function[i])
gen_fun(['Yashwanth','Sravanthi','Meenaskhi','Sarina'])
```

```python
# Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
x = [10,20,30,40,50]
fahrenheit_temperature = map(lambda x : x * 9/5 + 32,x)
my_set = set()
for i in fahrenheit_temperature:
  my_set.add(i)
my_set
```
